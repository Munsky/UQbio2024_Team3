```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'TrueData/Drug8_1000nM_DesignResults.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

```python
# Calculate the mean num_rna_cyto for each combination of Rep_Num and FOV_Num at each time point
rep_fov_mean_data = data.groupby(['Rep_Num', 'FOV_Num', 'Time'])['num_rna_cyto'].mean().reset_index()

# Calculate the overall mean num_rna_cyto for each time point
overall_mean_data = data.groupby('Time')['num_rna_cyto'].mean().reset_index()

# Plotting the data
plt.figure(figsize=(12, 6))

# Get unique combinations of Rep_Num and FOV_Num
rep_fov_combinations = rep_fov_mean_data[['Rep_Num', 'FOV_Num']].drop_duplicates()

# Plot each combination of Rep_Num and FOV_Num
for _, row in rep_fov_combinations.iterrows():
    rep_num = row['Rep_Num']
    fov_num = row['FOV_Num']
    subset_data = rep_fov_mean_data[(rep_fov_mean_data['Rep_Num'] == rep_num) & (rep_fov_mean_data['FOV_Num'] == fov_num)]
    label = f"Rep {rep_num}, FOV {fov_num}"
    plt.plot(subset_data['Time'], subset_data['num_rna_cyto'], label=label, alpha=0.7)

# Plot the overall mean line
plt.plot(overall_mean_data['Time'], overall_mean_data['num_rna_cyto'], label='Overall Mean', color='black', linewidth=2, linestyle='--')

plt.xlabel('Time')
plt.ylabel('Number of RNA in Cytoplasm')
plt.title('Mean RNA Count in Cytoplasm Over Time by Rep and FOV with Overall Mean')
plt.legend(title='Rep and FOV', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
