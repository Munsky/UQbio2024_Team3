```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os

# Function to process each file and compute the mean mRNA counts per frame
def process_file(file_path1, file_path2):
    data = pd.read_csv(file_path1)
    # Read the life_span column from Basic_Information
    life_span = pd.read_csv(file_path2)['life_span']
    # Initialize the 0/1 matrix
    matrix = np.zeros((len(life_span), 21))

    # Fill in the matrix
    for i, span in enumerate(life_span):
        matrix[i, :int(span)] = 1

    # Convert life_span to a DataFrame
    life_span_df = pd.DataFrame(matrix)

    # Convert data.iloc[:, 1:] and life_span_df to NumPy arrays
    data_array = data.iloc[:, 1:].values
    life_span_array = life_span_df.values

    # Multiply data_array with life_span_array
    product_array = data_array * life_span_array

    # Sum each column
    sum_product = product_array.sum(axis=0)

    # Sum each column in life_span
    sum_life_span = life_span_df.sum(axis=0)

    # Divide sum_product by sum_life_span to get the mean values
    mean_values = sum_product / sum_life_span
    return mean_values

# List to store the mean values of each file
mean_values_list = []
file_labels = []

# Iterate over each file and calculate mean mRNA counts
for i in range(1, 3): # Adjust the range according to your file naming
    for j in range(1, 7): # Adjust the range according to your file naming
        file_path = f'NoDrug/rep{i}_FOV{j}/mRNAs_in_nucleus.csv'
        file_path2 = f'NoDrug/rep{i}_FOV{j}/Basic_Information.csv'
        if os.path.exists(file_path):
            mean_values = process_file(file_path,file_path2)
            mean_values_list.append(mean_values)
            file_labels.append(f'rep{i}_FOV{j}')

# Combine all mean values into a DataFrame
mean_values_df = pd.DataFrame(mean_values_list, index=file_labels)

# Calculate the mean of all files
overall_mean = mean_values_df.mean()

# Plotting
plt.figure(figsize=(12, 8))

# Plot each file's mean values
for file_label in mean_values_df.index:
    plt.plot(range(1, len(mean_values_df.columns) + 1), mean_values_df.loc[file_label], label=file_label)

# Plot the overall mean
plt.plot(range(1, len(overall_mean) + 1), overall_mean, label='Overall Mean', linewidth=2, linestyle='--', color='black')

# Adding title and labels
plt.title('Mean mRNA Counts in Nucleus Over Time')
plt.xlabel('Frame Number')
plt.ylabel('Mean mRNA Count')

# Adding a legend
plt.legend(title='File', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.show()
```
