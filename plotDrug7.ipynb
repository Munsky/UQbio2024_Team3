## mRNA Counts in Nucleus Over Time for Drug 7

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

# Function to process each file and compute the mean mRNA counts per frame
def process_file(file_path):
    data = pd.read_csv(file_path)
    mean_values = data.iloc[:, 1:].mean()
    return mean_values

# List to store the mean values of each file
mean_values_list = []
file_labels = []

# Iterate over each file and calculate mean mRNA counts
for i in range(1, 3): # Adjust the range according to your file naming
    for j in range(1, 7): # Adjust the range according to your file naming
        file_path = f'Drug7_1000nM/rep{i}_FOV{j}/mRNAs_in_nucleus.csv'
        if os.path.exists(file_path):
            mean_values = process_file(file_path)
            mean_values_list.append(mean_values)
            file_labels.append(f'rep{i}_FOV{j}')

# Combine all mean values into a DataFrame
mean_values_df = pd.DataFrame(mean_values_list, index=file_labels)

# Calculate the mean of all files
overall_mean = mean_values_df.mean()

# Plotting
plt.figure(figsize=(12, 8))

# Plot each file's mean values
for file_label in mean_values_df.index:
    plt.plot(range(1, len(mean_values_df.columns) + 1), mean_values_df.loc[file_label], label=file_label)

# Plot the overall mean
plt.plot(range(1, len(overall_mean) + 1), overall_mean, label='Overall Mean', linewidth=2, linestyle='--', color='black')

# Adding title and labels
plt.title('Mean mRNA Counts in Nucleus Over Time')
plt.xlabel('Frame Number')
plt.ylabel('Mean mRNA Count')

# Adding a legend
plt.legend(title='File', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.show()
```
